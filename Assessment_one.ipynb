Question 11

```python
q1 = df[["Item", "Y2014", "Y2017"]]
q11 = q1.groupby(by = "Item").agg({'Y2014': 'sum', 'Y2017': 'sum'}).reset_index()
#specify the columns to sum
cols = ['Y2014', 'Y2017']

#find sum of columns specified 
q11['sum'] = q11[cols].sum(axis=0)
q11
```
